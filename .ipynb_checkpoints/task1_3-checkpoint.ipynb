```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import math

%matplotlib inline

r_data = pd.read_csv('Dataset .csv')
```

```python
r_data.describe()
```

```python
r_data.isnull().any().count()
```

```python
r_data = r_data.dropna()
```

```python
r_data.isnull().any()
```

```python
r_data
```

```python
r_data_refined = r_data.drop(columns=['Restaurant ID','Restaurant Name','Is delivering now','Rating text','Rating color'], axis=1)
```

```python
r_data_refined
```

```python
r_data_refined['Country Code'].unique()
```

```python

from sklearn.preprocessing import OneHotEncoder
encoder = OneHotEncoder(handle_unknown='ignore', sparse_output = False).set_output(transform='pandas')
```

```python
encodertransform = encoder.fit_transform(r_data_refined[['Country Code','Currency','Has Table booking','Has Online delivery','Switch to order menu',]])
```

```python
r_data_refined = pd.concat([r_data_refined, encodertransform], axis=1).drop(columns = ['Country Code','Currency','Has Table booking','Has Online delivery','Switch to order menu'])
```

```python
from sklearn.preprocessing import LabelEncoder

city_encoder = LabelEncoder()
r_data_refined['city_label'] = city_encoder.fit_transform(r_data_refined['City'])
```

```python
r_data_refined
```

```python
r_data_refined = r_data_refined.drop(columns=['Address','Locality','Locality Verbose'], axis=1)
```

```python
r_data_refined
```

```python
from sklearn.preprocessing import MultiLabelBinarizer
r_data_refined['Cuisines'] = r_data_refined['Cuisines'].apply(lambda x: x.split(", "))
```

```python
mlb = MultiLabelBinarizer()
cuisine_encoded = mlb.fit_transform(r_data_refined['Cuisines'])
cuisine_encoded_data = pd.DataFrame(cuisine_encoded, columns=mlb.classes_)
r_data_refined = pd.concat([r_data_refined, cuisine_encoded_data], axis=1)
```

```python
r_data_refined.head()
```

```python
r_data_refined = r_data_refined.drop(columns=['City','Cuisines'], axis=1)
```

```python
r_data_refined.dropna()
```

```python
corr=r_data_refined.corr()
sns.heatmap(corr,annot=True,cmap='coolwarm')
```
